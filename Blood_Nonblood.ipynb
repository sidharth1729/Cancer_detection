```python
from keras.layers import Dense,Flatten,Conv2D,MaxPooling2D,Activation,Dropout,Input,BatchNormalization,add
from keras.models import Sequential,Model
from keras.preprocessing.image import ImageDataGenerator
```

```python
batchsize = 10
target_size=(300, 300)
data_generator = ImageDataGenerator(rescale=1.0/255)
train_data = data_generator.flow_from_directory('/kaggle/input/TrainingImagesold',target_size=target_size,batch_size=batchsize,class_mode='binary')
test_data = data_generator.flow_from_directory('/kaggle/input/TestingImagesold',target_size=target_size,batch_size=batchsize,class_mode='binary')
```

```python
#Alexnet model
model = Sequential()
model.add(Conv2D(32, (3, 3), strides = 1, padding = "same", activation = "relu", input_shape = (300, 300,3)))
model.add(Dropout(0.4))
model.add(Conv2D(32, (3, 3), strides = 1, padding = "same", activation = "relu"))
#model.add(Dropout(0.4))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.3))

model.add(Conv2D(64, (3, 3), strides = 1, padding = "same", activation = "relu"))
model.add(Dropout(0.4))
model.add(Conv2D(64, (3, 3), strides = 1, padding = "same", activation = "relu"))
#model.add(Dropout(0.4))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.4))

model.add(Conv2D(128, (3, 3), strides = 1, padding = "same", activation = "relu"))
model.add(Dropout(0.4))
model.add(Conv2D(128, (3, 3), strides = 1, padding = "same", activation = "relu"))
#model.add(Dropout(0.4))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(1, activation = "sigmoid"))
```

```python
#mod=model.fit_generator(train_data,epochs=50,validation_data=test_data,verbose=1,steps_per_epoch=12,validation_steps=12,shuffle=True,callbacks=[keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)])
model.compile(loss='binary_crossentropy',optimizer=keras.optimizers.Adam(1e-3),metrics=['accuracy'])
mod=model.fit_generator(train_data,epochs=5,validation_data=test_data,verbose=1,steps_per_epoch=12,validation_steps=12,shuffle=True,callbacks=[keras.callbacks.EarlyStopping(monitor='val_loss',patience=5)])
```

```python
epochs= range(1,23)

y_ticks = [i/10 for i in range(1,11)]

plt.plot(epochs,mod.history['loss'],label='loss')
plt.plot(epochs,mod.history['val_loss'],label='val_loss')
#plt.xticks(epochs)
#plt.yticks(y_ticks)
plt.legend()
plt.show()
```

```python
plt.plot(epochs,mod.history['accuracy'],label='accuracy')
plt.plot(epochs,mod.history['val_accuracy'],label='val_accuracy')
#plt.xticks(epochs)
#plt.yticks(y_ticks)
plt.legend()
plt.show()
```
